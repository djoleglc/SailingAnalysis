```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
sailing_class_dict = {
    "1": "IQfoil men",
    "2": "IQfoil women",
    "3": "Kitefoil men",
    "4": "Kitefoil women",
    "5": "49er",
    "6": "49er FX",
    "7": "470 Mix",
    "8": "ILCA 6",
    "9": "ILCA 7",
    "10": "Nacra17"
}


def create_summary_athlete():
    df_mod = df
    df_mod["Pathology"] = 1
    result = df_mod.groupby('ID').agg({
    'Pathology': 'sum',
    'Sex': "mean",
    'Sailing Class': 'mean', 
    'International regata': 'nunique'
    }).reset_index()
    
    result.rename(columns = {"Pathology": "N. Injuries"}, inplace=True)
    result.rename(columns = {"International regata": "Total Events attended"}, inplace=True)

    result["Sex"] = result["Sex"].apply(lambda x : "Male" if x==1 else "Women")
    result["Sailing Class"] = result["Sailing Class"].apply(lambda x : sailing_class_dict[str(int(x))])
    result["Avg Injuries per Regata"] = np.round(result["N. Injuries"]/result["Total Events attended"],2)
    
    
    result.set_index("ID", inplace=True)
    
    return result

df_athletes = create_summary_athlete()
print(df_athletes.to_latex())
```

```python
pathology_dict = {
    "1": "Concussion",
    "2": "Fracture (traumatic)",
    "3": "Stress fracture (overuse)",
    "4": "Dislocation, subluxation",
    "5": "Ligamentous rupture",
    "6": "Sprain (joint and/or ligament)",
    "7": "Lesion of meniscus or cartilage",
    "8": "Strain/muscle injury",
    "9": "Contusion/haematoma",
    "10": "Tendinosis/tendinopathy",
    "11": "Fasciitis/aponeurosis injury",
    "12": "Spinal cord injury",
    "13": "Muscle cramps",
    "14": "Compartimental syndrome",
    "15": "Acute overload"
}


def injuries_analysis():
    return df["Pathology"].astype(str).apply(lambda x: pathology_dict[x])
```

```python
freq = injuries_analysis().value_counts()
print((freq/sum(freq)).to_latex())
```
